```python

scaled_df=scalar.fit_transform(df)
```

DIMENSIONALITY REDUCTION

```python
pca=PCA(n_components=2)
principal_component=pca.fit_transform(scaled_df)
pca_df=pd.DataFrame(data=principal_component,columns=['PCA1','PCA2'])
pca_df
```

FINDING THE VALUE OF K USING ELBOW  METHOD

```python
inertia=[]
range_val=range(1,15)
for i in range_val:
  kmeans=KMeans(n_clusters=i)
  kmeans.fit_transform(pd.DataFrame(scaled_df))
  inertia.append(kmeans.inertia_)
plt.plot(range_val,inertia,'bx-')
plt.xlabel('Value of k')
plt.ylabel('Inertia')
plt.title('the elbow method using inertia')
plt.show()


```

inertia_ is basically the minimum distance of sum of square of the distance from the centroid  , here the inertia[] and inertia_ is different...........

```python
kmeans_model=KMeans(4)
ax=kmeans_model.fit_transform(scaled_df)
#pd_df_kmeans=pd.concat([pca_df,pd.DataFrame({'cluster':kmeans_model.labels_})],axis=1)
ax
```

The kmeans_model.labels_ attribute in the KMeans model stores the cluster labels for each data point after fitting the model.

```python
kmeans_model=KMeans(4)
ax=kmeans_model.fit_transform(scaled_df)
pd_df_kmeans=pd.concat([pca_df,pd.DataFrame({'cluster':kmeans_model.labels_})],axis=1)
```

TRAINING AND TESTING THE MODEL ACCURACY USING DECISION TREE

```python
x=cluster_df.drop(['Cluster'],axis=1)
y=cluster_df['Cluster']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
model=DecisionTreeClassifier(criterion="entropy")
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
```

```python
y_pred
```

CONFUSION_MATRIX

```python
#confusion_matrix
print(metrics.confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))

```

SAVING THE DECISION TREE MODEL FOR THE FUTURE PREDICTION

```python
import pickle
filename ='final_model.sav'
pickle.dump(model,open(filename,'wb'))

#some time later...........

# load the model from disk

loaded_model = pickle.load(open(filename, 'rb'))
result = loaded_model.score(x_test, y_test)
print(result,'% Acuuracy')

```
